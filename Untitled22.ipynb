```python
airbnb.drop(['id','host_name','last_review'],axis=1,inplace=True)
airbnb.head()
```

```python
airbnb['reviews_per_month'].fillna(0,inplace=True)
```

```python
top_host_id = airbnb['host_id'].value_counts().head(10)
print(top_host_id)
```

```python
#138649185 is the top_host_id
```

```python
sns.set(rc={'figure.figsize':(10,8)})
viz_bar = top_host_id.plot(kind='bar')
viz_bar.set_title('Hosts with the most listings in Singapore')
viz_bar.set_xlabel('Host IDs')
viz_bar.set_ylabel('Count of listings')
viz_bar.set_xticklabels(viz_bar.get_xticklabels(), rotation=45)
plt.savefig('top_host_listings.png')

# Show the plot (optional)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=airbnb, x='room_type')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
plt.savefig('Room Type Distribution.png')
```

```python
# private room is the common type
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'airbnb' with the 'neighbourhood_group' column
if airbnb['neighbourhood_group'].isna().all():
    print("No information available for 'neighbourhood_group'")
else:
    labels = airbnb['neighbourhood_group'].value_counts().index
    colors = ['#008fd5', '#fc4f30', '#e5ae38', '#6d904f', '#8b8b8b']
    explode = (0.1, 0, 0, 0, 0)
    shape = airbnb['neighbourhood_group'].value_counts().values
    plt.figure(figsize=(12, 12))
    plt.pie(shape, explode=explode, labels=shape, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.legend(labels)
    plt.title('Neighbourhood Group')
    plt.savefig('2.png')
    plt.show()
```

```python
#  central region has the most listing
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'airbnb_1' with the 'neighbourhood_group' and 'price' columns
if airbnb['neighbourhood_group'].isna().all():
    print("No information available for 'neighbourhood_group'")
else:
  plt.style.use('classic')
  plt.figure(figsize=(13,7))
  plt.title("Neighbourhood Group Price Distribution < 300")
  sns.boxplot(y="price",x ='neighbourhood_group' ,data = airbnb[airbnb.price<300])
  plt.show()
  plt.savefig('boxplot.png')
```

```python
#  central region has medium price of 150
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = airbnb.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
plt.savefig('heatmap.png')
# # - Strong positive correlation between 'review_per_month' and 'number_of_review' and between 'review_per_month' and 'number_of_review_ltm'
```

```python
# Assuming you have a DataFrame named 'Listing' with the 'availability_365' column
import plotly.express as px
if airbnb['neighbourhood_group'].isna().all():
    print("No information available for 'neighbourhood_group'")
else:
    fig = px.pie(airbnb, names='neighbourhood_group', values='availability_365')
    fig.show()
    plt.savefig('piechart.png')
```
