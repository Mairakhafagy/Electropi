```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
df
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.pairplot(df, hue='Species')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x= df.drop('Species',axis= 1)
y = df['Species']
```

```python
x_scl = StandardScaler().fit_transform(x)
```

```python
x_scl
```

```python
from sklearn.decomposition import PCA
```

```python
dim_red = PCA()
```

```python
x_pca = dim_red.fit_transform(x_scl)
```

```python
x_pca.shape
```

```python
dim_red.components_
```

```python
dim_red.explained_variance_ratio_
```

```python
dim_red.get_feature_names_out()
```

```python
df_pca = pd.DataFrame(x_pca,columns = dim_red.get_feature_names_out())
```

```python
df_pca['Species']= y
```

```python
sns.pairplot(df_pca,hue= 'Species') 
```
